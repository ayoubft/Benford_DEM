```python
# IMPORTING LIBRARIES
import numpy as np
import matplotlib.pyplot as plt
import rasterio
from LNC import get_dln, f_lnd, fdln
```

```python
# Load DEM
dem = rasterio.open('tifs/MNT_carriere.tif')

# Load aspect (slope)
asp0 = rasterio.open('tifs/MNT_carriere_aspect0.tif')

asp1 = rasterio.open('tifs/MNT_carriere_aspect_trigo.tif')
```

```python
print("\nDEM\n")
print(dem.profile)
print("\nASPECT 0\n")
print(asp0.profile)
print("\nASPECT 1\n")
print(asp1.profile)
```

```python
# reading the first band (not from zero!)
band1 = dem.read(1)
```

```python
# Preview of the 
band1
```

```python
band1 = band1.flatten()
```

```python
f1 = get_dln(band1)
```

```python
n = sum(f1)
ff1 = []
for i in range(len(f1)):
    ff1.append(100*f1[i]/n)
```

```python
f_lnd(f1)
```

```python
# reading the first band (not from zero!)
band2 = asp0.read(1) # 2d array
```

```python
band2 = band2.flatten()
```

```python
f2 = get_dln(band2)
```

```python
n2 = sum(f2)
ff2 = []
for i in range(len(f2)):
    ff2.append(100*f2[i]/n2)
```

```python
f_lnd(f2)
```

```python
# reading the first band (not from zero!)
band3 = asp1.read(1) # 2d array
```

```python
band3 = band3.flatten()
```

```python
f3 = get_dln(band3)
```

```python
n3 = sum(f3)
ff3 = []
for i in range(len(f3)):
    ff3.append(100*f3[i]/n2)
```

```python
f_lnd(f3)
```

```python
X = list(range(1,  10))

B = [30.1, 17.6, 12.5, 9.7, 7.9, 6.7, 5.8, 5.1, 4.6]
```

```python
# Plotting both the curves simultaneously
plt.figure(figsize=(15, 12))
plt.plot(X, B, color='k', label='Benford')
plt.plot(X, ff1, color='r', label='DEM')
plt.plot(X, ff2, color='g', label='ASPECT 0')
plt.plot(X, ff3, color='y', label='ASPECT 1')
  
# Naming the x-axis, y-axis and the whole graph
plt.xlabel("Leading Digit")
plt.ylabel("%")
plt.title("Leading Digits Distributions")
  
# Adding legend, which helps us recognize the curve according to it's color
plt.legend()
  
# To load the display window
plt.show()
```

```python
N = 9
ind = np.arange(N) 
width = 0.2

plt.figure(figsize=(15, 12))

bar1 = plt.bar(ind, B, width, color = 'r')

bar2 = plt.bar(ind+width, ff1, width, color='g')
  
bar3 = plt.bar(ind+width*2, ff2, width, color = 'b')

bar4 = plt.bar(ind+width*3, ff3, width, color = 'k')
  
plt.xlabel("Digits")
plt.ylabel('%')
plt.title("Leading Digits Count")
  
plt.xticks(ind+width,['1', '2', '3', '4', '5', '6', '7', '8', '9'])
plt.legend( (bar1, bar2, bar3, bar4), ('Benford', 'DEM', 'ASPECT', 'ASPECT 1') )
plt.show()
```
